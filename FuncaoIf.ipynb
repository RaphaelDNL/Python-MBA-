Treinando com input e if, else e elif

```python
#é estudante?
aluno = input("Olá! Você é estudante, SIM ou Não? ")
nome = input("Digite seu nome: ")

#Nota do estudante
top = 10
media = 8
pessimo = 3

if nome == "Lilian" or nome == "Marcos":
  print("Parabéns, voce tirou nota: ", top)
elif nome == "Luan" or nome == "Raphael":
  print("Parabéns, você tirou nota: ", media)
else:
  print("É melhor estudar mais, você tirou nota: ", pessimo)
```

```python
#Lista de estudantes com suas notas
notas_estudantes = {
    "Lilian": 10,
    "Raphael": 8,
    "Marcos": 7,
    "Luan": 3,
    "Joãa": 0,
}

#perguntando o nome
nome = input("Digite seu nome: ")

#If na lista de alunos com base no nome
if nome in notas_estudantes: #o comando in pesquisa dentro da lista
  nota = notas_estudantes[nome]
  print("Parabéns, sua nota é ", nota)
else:
  print("Não localizado")
```

```python
#Cadastro de carros
lista_carros = {
    "ka": "Ford",
    "Fusca": "VW",
    "Celta": "Chevrolet",
    "Sandeiro": "Renault",
}


#Input qual modelo de carro é
carro = input("Digite seu o modelo do seu carro aqui: ")

#If na lista de cadastro com base no input carro
if carro in lista_carros:
  modelo = lista_carros[carro]
  print("Seu carro é da marca", modelo)
else:
  print("Marca não localizada")
```

```python
#cadastro de nomes e aniversário
lista_mebros = {
    "Raphael": "29/09",
    "Lilian": "04/05",
    "Regina": "31/08",
    "Alexandre": "01/10",
}
#aniversário de quem procura
aniversariante = input("Quem é o aniversáriante? ")
#retornar a data de aniversário
if aniversariante in lista_mebros:
  nome = lista_mebros[aniversariante]
  print("O aniversário do(a) ", aniversariante, "é em ", nome)
else:
  print("Não encontrado, favor cadastrar pessoa como mebro da igreja")
```
